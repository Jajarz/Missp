```python
from pulp import *
```

```python
prob = LpProblem("Produkcja problem",LpMaximize)
sztukA=LpVariable("sztukA",0,None,LpInteger)
sztukB=LpVariable("sztukB",0)
sztukC=LpVariable("sztukC",0)


prob += 400*sztukA + 300*sztukB +200*sztukC, "Optymalizacja zysku"
prob += 0.3*sztukA + 0.5*sztukB+0.4*sztukC <= 800, "Montaż godzin"
prob += 0.1*sztukA + 0.08*sztukB+0.04*sztukC <= 800, "Kontrola godzin"
prob += 0.06*sztukA + 0.04*sztukB+0.05*sztukC <= 700, "Pakowanie godzin"
prob.writeLP("Kontrola.lp")
prob.solve()

for v in prob.variables():
    print(v.name, "=", v.varValue)
print("Całkowity zysk = ", value(prob.objective))
```

```python
prob = LpProblem("Problem kota",LpMinimize)
x1=LpVariable("Chicken",0,None,LpInteger)
x2=LpVariable("Beef",0)
x3=LpVariable("Mutton",0)
x4=LpVariable("Rice",0)
x5=LpVariable("Wheat bran",0)
x6=LpVariable("Gel",0)
x7=LpVariable("Sardines",0)

# liczenie

prob += 0.013*x1 + 0.008*x2 + 0.010*x3 + 0.002*x4 + 0.005*x5 + 0.001*x6 + 0.004*x7, "Cena końcowa"
prob += x1+x2+x3+x4+x5+x6+x7 == 100, "Ilość jednostek"
prob += 0.100*x1 + 0.200*x2 + 0.150*x3 + 0.000*x4 + 0.040*x5 + 0.000*x6 + 0.020*x7 >= 8.0, "Ilość protein"
prob += 0.080*x1 + 0.100*x2 + 0.110*x3 + 0.010*x4 + 0.010*x5 + 0.000*x6 + 0.030*x7 >= 6.0, "Ilośc tłuszczu"
prob += 0.001*x1 + 0.005*x2 + 0.003*x3 + 0.100*x4 + 0.150*x5 + 0.000*x6 + 0.200*x7 <= 2.0, "Ilość błonnika"
prob += 0.002*x1 + 0.005*x2 + 0.007*x3 + 0.002*x4 + 0.008*x5 + 0.000*x6 + 0.015*x7 <= 0.4, "Ilość soli"
prob += 0.030*x1 + 2.600*x2 + 2.600*x3 + 0.000*x4 + 0.000*x5 + 0.000*x6 + 8.900*x7 >= 0.8, "Ilość witB"

prob.writeLP("Kontrola1.lp")
prob.solve()

#for v in prob.variables():
#    print(v.name, "=", v.varValue)
#print("Całkowity zysk = ", value(prob.objective))

#sklejone_produkt_cena = [":".join((produkt, str(cena))) for produkt,cena in zip(["chicken","beef","mutton","rice","wheat","gel","sardines"],[0.013,])]
print("Końcowy produkt będzie kosztować ", value(prob.objective), "centów, będzie miał w składzie: ")

for v in prob.variables():
    print(v.name, ":", v.varValue, "%, ")
```
