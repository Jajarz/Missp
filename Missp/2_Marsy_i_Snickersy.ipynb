```python
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual, Layout, FloatSlider, IntSlider
from ipywidgets import FloatSlider
import pandas as pd

style = {'description_width': 'initial'}

cena_czekolady_slider = FloatSlider(min=0,max = 100, step=0.1, value= 20.0, description="Cena czekolady" )
cena_orzechow_slider = FloatSlider(min=0,max = 100, step=0.1, value= 10.0, description="Cena orzechow" )

def hurt_produkcja(cena_czekolady= cena_czekolady_slider,
cena_orzechow = cena_orzechow_slider):

  prob1 = LpProblem("Problem produkcji",LpMaximize)
  Cz=LpVariable("Czekolada",0,None,LpInteger)
  K=LpVariable("Karmel",0)
  Cu=LpVariable("Cukier",0)
  Ol=LpVariable("Olej",0)
  Orz=LpVariable("Orzechy",0)

  Cm=LpVariable("Czekolada na marsa",0,None,LpInteger)
  Cs=LpVariable("Czekolada na snickersa",0)

  S=LpVariable("Snickers",0)
  M=LpVariable("Mars",0)
  prob1 += S + M <= 10000, "Maksymalna ilosc produktow"
  prob1 += Orz <= 3000, "Ile orzechow na snickersa"
  prob1 += Cz <= 6000, "Ilosc czekolady lacznie"



  prob1 += 5*Orz == S, "Ilosc snicekrsow - orzechy"
  prob1 += 2.5*Cs == S, "Ilosc snicekrsow - czekolada"
  prob1 += 2*Cm == M, "Ilosc marsow"
  prob1 += Cm + Cs == Cz, "Ilosc czekolady na batony"
  prob1 += 100*M + 105*S, "Zysk"
  prob1 += 100* M + 105* S - cena_czekolady * Cz - cena_orzechow * Orz, "Koszt"
  prob1.writeLP("Kontrola2.lp")
  prob1.solve()
  for v in prob1.variables():
      print(v.name, ":", v.varValue, "kg, ")
  print("Ca≈Çkowity zysk = ", value(prob1.objective))

  # shadows = [{'name':name, 'shadow price': c.pi, "slack": c.slack} for name, c in prob1.constraints.items()]
  #   print("*"*10+"Shadow Prices"+"*"*10)
  #   print(pd.DataFrame(shadows))

interact(hurt_produkcja,
         cena_czekolady=cena_czekolady_slider,
         cena_orzechow=cena_orzechow_slider)
```
